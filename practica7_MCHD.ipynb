El objetivo de esta práctica es  cargar y explorar un dataset, y luego utilizar matplotlib para crear gráficos visuales que permitan analizar e interpretar los datos.

## Dataset:

Para este ejercicio, utilizaremos el famoso dataset `Iris` de la librería `seaborn`. Este dataset contiene información sobre diferentes especies de flores Iris y varias medidas de sus características (como el largo y ancho del sépalo y el pétalo).

### Carga de Dataset

Importamos la librería `seaborn` y matplotlib para utilizar el dataset `Iris` y realizar los gráficos

```python
import seaborn as sns
import matplotlib.pyplot as plt

# carga del dataset
df = sns.load_dataset('iris')

# Impresión de las primeras filas
print(df.head())

# Obtener la información gral del dataset
print(df.info())
```

## Visualización de Datos

### Gráfico de barras

**Objetivo:** Comparar el promedio de largo y ancho de los pétalos entre las tres especies de Iris.

**Instrucciones:** Utiliza un gráfico de barras para mostrar la media de las columnas `petal_length` y `petal_width` agrupadas por species.

```python
# .rename cambia el nombre de las etiquetas de x y de y en la gráfica
df.groupby('species')[['petal_length', 'petal_width']].mean().rename(columns={'petal_length': 'Largo de pétalo', 'petal_width': 'Ancho de pétalo'}).plot(kind='bar', color=['maroon','navy'])

plt.title('Promedio de Largo y Ancho del Sépalo por Especie')
plt.xlabel('Especie')
plt.ylabel('Medida(cm)')
plt.xticks(rotation=0) # para mostrar las etiquetas de x de manera horizontal
plt.legend(title='Tipo')
plt.grid(axis='y', linestyle='--', alpha=0.5) # genera líneas de referencia en y
plt.show()
```

### Histograma

**Objetivo:** Visualizar la distribución de los largos de los sépalos (`sepal_length`).

**Instrucciones:** Crea un histograma de la variable `sepal_length` y asegúrate de que tenga un número adecuado de bins. Coloca etiquetas en los ejes y un título.

```python
plt.hist(df['sepal_length'], bins=20, color='firebrick', edgecolor='black')
plt.xlabel('Largo del Sépalo(cm)')
plt.ylabel('Frecuencia')
plt.title('Distribución del Largo del Sépalo')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

## Gráfico de dispersión (scatter plot):

**Objetivo:** Relacionar el largo del pétalo con el ancho del pétalo.

**Instrucciones:** Crea un gráfico de dispersión que compare `petal_length` con `petal_width`. Usa un color diferente para cada especie.

```python
# Utilizamos scatterplot de seaborn

plt.figure(figsize=(15, 6))

# hue se usa para diferenciar cada especie por colores
# s lo utilizamos para aumentar o disminuir el tamaño de los puntos
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', palette=['maroon', 'navy', 'green'], s=75)

plt.title('Relación entre Largo y Ancho del Pétalo')
plt.xlabel('Largo del Pétalo(cm)')
plt.ylabel('Ancho del Pétalo(cm)')
plt.legend(title='Especie')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.5)# genera líneas de referencia en x
plt.show()
```

## Gráfico de cajas (box plot):

**Objetivo:** Analizar la distribución de las medidas del sépalo (`sepal_length`, `sepal_width`).

**Instrucciones:** Crea un gráfico de cajas para mostrar la distribución de las variables `sepal_length` y `sepal_width` por especie.

```python
# Gráfico de cajas para sepal_length
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='species', y='sepal_length', hue='species', palette='pastel')
plt.title('Distribución de la longitud\ndel sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Longitud del sépalo(cm)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Gráfico de cajas para sepal_width
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='species', y='sepal_width', hue='species', palette='viridis')
plt.title('Distribución del ancho\ndel sépalo por especie')
plt.xlabel('Especie')
plt.ylabel('Ancho del Sépalo(cm)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

## Gráfico de líneas:

**Objetivo:** Mostrar la tendencia de la longitud del pétalo (`petal_length`) a través de las observaciones.

**Instrucciones:** Crea un gráfico de líneas utilizando la variable `petal_length` (en el eje Y) con el índice de las observaciones (en el eje X). Puedes hacer esto para ver cómo varía la longitud del pétalo a lo largo del dataset.

```python
plt.figure(figsize=(20, 6)) # Aumentar el tamaño de la figura para que sean más visibles los datos
plt.plot(df.index, df['petal_length'], marker='o', linestyle='-', color='steelblue', alpha=0.6)
plt.title('Tendencia de la Longitud del Pétalo')
plt.xlabel('Índice de Observación')
plt.ylabel('Longitud del Pétalo(cm)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```
