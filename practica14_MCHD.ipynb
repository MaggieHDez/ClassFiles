## **Ejercicio 01. Errores Tipográficos y Valores Inconsistentes: maritalstatus**

### **Instrucciones:**
1. Imprime de la columna `'maritalstatus'` los valores únicos.
2. Imprime la frecuencia de cada valor único
3. Cambia el valor `'Never-Married'` a `'Single'`.
4. Imprime la frecuencia de cada valor único.

```python
# Importamos pandas
import pandas as pd
# Leemos el archivo
archivo = 'drive/MyDrive/ClassFiles/nls97.csv'
df = pd.read_csv(archivo, index_col=0)

df.head()
```

```python
# Imprimimos los valores unicos de maritalstatus
print('Columna maritalstatus\n', df['maritalstatus'].unique())

print('\n********************************************************************\n')

# Imprimimos la frecuencia de cada valor único
print('Frecuencia de cada valor único\n', df['maritalstatus'].value_counts())

print('\n********************************************************************\n')

# Reemplazamos Never-married a Single
df['maritalstatus'] = df['maritalstatus'].replace('Never-married', 'Single')
# Imprimimos nuevamente los valores únicos
print('Frecuencia de cada valor único\n', df['maritalstatus'].value_counts())
```

## **Ejercicio 02. Errores Tipográficos y Valores Inconsistentes: highestdegree**
### **Instrucciones:**
1. Imprime de la columna `'highestdegree'` los valores únicos.
2. Imprime la frecuencia de cada valor único
3. Observa que cada valor categórico empieza con 0. Modifica las categorías para que todas estén en minúsculas y ninguna empiece con un número. Por ejemplo, si dice `"2. High School"`, debería ser `'high school'`.
4. Los cambios se realizarán en la misma columna `'highestdegree'`.
5. Imprime los nuevos valores únicos.

```python
# Imprimimos los valores unicos de highestdegrees
print('Columna highestdegree\n', df['highestdegree'].unique())

print('\n********************************************************************\n')

# Imprimimos la frecuencia de los valores únicos de highestdegree
print('Frecuencia de cada valor único\n', df['highestdegree'].value_counts())
```

```python
# Reemplazamos los números, punto y espacio al inicio de la categoría y
# convertimos en minúsculas
df['highestdegree'] = df['highestdegree'].str.lower().replace(r'^\d+\.?\s*', '', regex=True)

# Imprimimos los valores unicos de highestdegrees
print('Columna highestdegree\n', df['highestdegree'].unique())
```

## **Ejercicio 03. Alta Cardinalidad - Frecuencias Bajas**
### **Instrucciones:**
1. De la columna `'maritalstatus'` une las categorías `'Widowed'` y `'Single'` en una categoría más general llamada `'Single/No Partner'`. Crea una nueva columna, `'marital_grouped'` donde aparezcan las nuevas categorías de maritalstatus.
2. Imprime los valores únicos de la nueva columna `'marital_grouped'`.
3. Imprime la frecuencia de los valores únicos de la nueva columna `'marital_grouped'`

```python
print('Frecuencia de cada valor único en maritalstatus\n', df['maritalstatus'].value_counts())

print('\n********************************************************************\n')

# Reemplazamos los valores de Widowed/Single y los agregamos a marital_grouped
df['marital_grouped'] = df['maritalstatus'].replace(['Widowed', 'Single'], 'Single/No Partner')

# Imprimimos los valores unicos de marital_grouped
print('Columna marital_grouped\n', df['marital_grouped'].unique())

print('\n********************************************************************\n')

# Imprimimos la frecuencia de los valores únicos de marital_grouped
print('Frecuencia de cada valor único en marital_grouped\n', df['marital_grouped'].value_counts())
```

## **Ejercicio 04. Tipos de Datos**
### **Instrucciones:**
1. Imprime el tipo de datos de la columna `'gender'`.
2. Imprime la cantidad de bytes de dicha columna
3. Si la columna es de tipo `object`, modifica a `category`.
4. En caso de haber realizado la modificación del tipo de datos, imprime la cantidad de bytes que ocupa dicha columna con el nuevo tipo de datos

```python
# Imprimimos el tipo de datos de Gender
print('Tipo de datos de Gender\n', df['gender'].dtype)

# Imprimimos la cantidad de bytes de Gender
print('Cantidad de bytes de Gender\n', df['gender'].memory_usage(deep=True))

# Cambiamos el tipo de datos a category
if df['gender'].dtype == 'object':
  df['gender'] = df['gender'].astype('category')

print('\n********************************************************************\n')

# Imprimimos la cantidad de bytes de Gender después del cambio
print('Cantidad de bytes de Gender después del cambio\n', df['gender'].memory_usage(deep=True))

# Imprimimos el tipo de datos de Gender
print('Tipo de datos de Gender después del cambio\n', df['gender'].dtype)
```
