## Importación de librería `re` y ejercicios

```python
import re

texto = """El 10 de agosto de 2025, María Pérez envió un correo electrónico a la dirección maria.perez@email.com solicitando información sobre el producto X-200. Su número de teléfono es +34 612-345-678. Además, visitó la web https://www.ejemplo.com para consultar precios y disponibilidad.
El servidor registró el acceso desde la IP 192.168.1.15 a las 14:23:56, y se detectó un error en el log con código ERR500. En Twitter, el usuario @DataAnalyst comentó: "Los modelos de IA son el futuro #MachineLearning #IA2025". El totaal de la compra fue de $199.99, y el cliente pagó con tarjeta VISA terminada en 1234.
Por otro lado, el DNI de Juan García es 12345678Z y su código postal 28080. Para más información, puede llamar al 911 234 567 o visitar www.informacion.es. Recordamos que las feechas límite para inscripción son 2025-09-30 y 2025-10-15."""

print('********************* TEXTO ORIGINAL *********************\n',texto)
print('\n**********************************************************\n')

# 1. Extraer todas las direcciones de correo electrónico.
email = re.findall(r'[\w\._%+-]+@[\w.-]+\.[A-Za-z]{2,}', texto)
print('Direcciones de correo electrónico\n', email)

# 2. Extraer todas las URLs (con o sin protocolo).
website = re.findall(r'(https?://\S+|www\.\S+)', texto)
print('\nURLs (con o sin protocolo)\n',website)

# 3. Validar y extraer números de teléfono en formato internacional y nacional.
telefono = re.findall(r'(\+\d{2} \d{3}-\d{3}-\d{3}|\d{3} \d{3} \d{3})', texto)
print('\nTeléfono en formato internacional y nacional\n', telefono)

# 4. Extraer fechas en formato yyyy-MM-dd.
fecha = re.findall(r'\d{4}-\d{2}-\d{2}', texto)
print('\nFechas en formato yyyy-MM-dd\n', fecha)

# 5. Extraer códigos postales de 5 dígitos.
codigo_postal = re.findall(r'\b\d{5}\b', texto)
print('\nCódigos postales de 5 dígitos\n', codigo_postal)

# 6. Extraer DNI (8 dígitos seguidos de letra mayúscula).
dni = re.findall(r'\b\d{8}[A-Z]\b', texto)
print('\nDNI (8 dígitos seguidos de letra mayúscula)\n', dni)

# 7. Extraer menciones de Twitter.
mentions = re.findall(r'(?<!\w)@\w+', texto) # lookbehind negativo; coincide @\w+ pero solo si NO hay una letra, guión o dígito
print('\nMenciones de Twitter(X)\n', mentions)

# 8. Extraer hashtags.
hashtags = re.findall(r'(?<!\w)#\w+', texto)
print('\nHashtags\n', hashtags)

# 9. Extraer direcciones IP.
dirIP = re.findall(r'\b(?:\d{1,3}\.){3}\d{1,3}\b', texto)
print('\nDirecciones IP\n', dirIP)

# 10. Extraer importes monetarios (con $ y decimales).
importe = re.findall(r'\$\d+\.\d{2}', texto)
print('\nImportes monetarios (con $ y decimales)\n', importe)

# 11. Eliminar letras repetidas.
# Si hay vocales repetidas 2 o más veces, remover 1
vocales = r'([aeiouAEIOU])\1+' # podemos agregar acentos y diéresis
texto_limpio = re.sub(vocales, r'\1', texto)

# Si hay consonantes repetidas 3 o más veces, remover 2
consonantes = r'([bcdfghjklmnpqrstvxyzBCDFGHJKLMNPQRSTVWXYZ])\1{2,}'
texto_limpio = re.sub(consonantes, r'\1', texto_limpio)

print('\n********************** TEXTO LIMPIO SIN LETRAS REPETIDAS **********************\n',texto_limpio)
print('\n*******************************************************************************\n')
```
