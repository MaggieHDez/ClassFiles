```python
# librerías
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import seaborn as sns
from sentence_transformers import SentenceTransformer
import numpy as np

nltk.download('stopwords', quiet = True)
nltk.download('wordnet', quiet = True)
nltk.download('punkt', quiet=True)

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

# Diccionario de modelos
modelos = {
    "Naive Bayes": MultinomialNB(),
    "SVM": SGDClassifier(loss="hinge", random_state=42),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
}

# Diccionarios de hiperparámetros para GridSearch
param_grid_nb = {
    "alpha": [0.1, 0.5, 1.0] # no es necesario para optimizar ya que "alpha" tiene poco impacto en rendimiento
}

param_grid_svm = {
    "loss": ["hinge"],
    "alpha": [0.0001, 0.001, 0.01]
}

param_grid_lr = {
    "C": [0.01, 0.1, 1, 10],
    "solver": ["lbfgs"],
    "max_iter": [500, 1000]
}

# Cargar modelo MiniLM (ligero y rápido)
model_mini = SentenceTransformer('all-MiniLM-L6-v2')

df = pd.read_csv("/content/drive/MyDrive/ClassFiles/PLN/DatasetSintomasMedicos.csv")
print(df.head(), '\n')

# EDA
def eda(df):
  print('Columnas: ', df.columns, '\n')

  print(df.info(), '\n')

  print(df.describe(include='all'), '\n')

  # shape
  print('\nShape: ', df.shape)

  # nulos
  print("\nNulos:\n", df.isnull().sum())

  # duplicados (en frase)
  print("\nDuplicados:", df.duplicated(subset=['phrase']).sum())

  # Eliminar duplicados
  df = df.drop_duplicates(subset=['phrase']).reset_index(drop=True)
  print("Nuevo shape después de eliminar duplicados:", df.shape)

  # Distribución de clases
  plt.figure(figsize=(12,6))
  df['prompt'].value_counts().plot(kind='bar')
  plt.title("Distribución de clases (Síntomas médicos)")
  plt.xticks(rotation=90)
  plt.tight_layout()
  plt.show()

  return df

# Preprocesamiento
def preprocesamiento(texto):
  texto = str(texto).lower() # Convertir a minúsculas
  texto = re.sub(r'<.*?>', '', texto)  # quitar HTML
  texto = re.sub(r'https?://\S+|www\.\S+', '', texto) # Eliminar URL
  texto = re.sub(r'[\w\._%+-]+@[\w.-]+\.[A-Za-z]{2,}', '', texto) # Eliminar email
  texto = re.sub(r'[^\w\s]', '', texto) # Eliminar puntuación
  texto = re.sub(r'\d+', '', texto)  # elimina números
  texto = re.sub(r'\s+', ' ', texto).strip() # Eliminar espacios
  # tokenizar
  tokens = texto.split()
  # quitar stopwords y lematizar
  tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
  return " ".join(tokens)

def entrenamiento(data):
  # Dividir los datos en conjuntos de entrenamiento y prueba
  X = data['cleaned']
  y = data['prompt']

  # TF-IDF
  tfidf = TfidfVectorizer()
  X_tfidf = tfidf.fit_transform(X)

  X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42, stratify=y)

  return X_train, X_test, y_train, y_test

def evaluar_modelo(nombre, modelo, X_train, X_test, y_train, y_test):
    modelo.fit(X_train, y_train)
    pred = modelo.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, pred)

    # Reporte por clase
    report = classification_report(y_test, pred, output_dict=True, zero_division=0)

    macro_prec = report["macro avg"]["precision"]
    macro_rec = report["macro avg"]["recall"]
    macro_f1 = report["macro avg"]["f1-score"]

    # Matriz de confusión normalizada
    cm = confusion_matrix(y_test, pred, normalize='true')

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=False, cmap="Blues")

    plt.title(f"Matriz de confusión normalizada - {nombre}")
    plt.ylabel("Etiqueta real")
    plt.xlabel("Predicción")

    plt.xticks(rotation=90, fontsize=6)
    plt.yticks(fontsize=6)
    plt.tight_layout()
    plt.show()



    return {
        "Modelo": nombre,
        "Accuracy": acc,
        "Precision_macro": macro_prec,
        "Recall_macro": macro_rec,
        "F1_macro": macro_f1
    }

def crear_wordcloud(tokens, titulo):
    if not tokens:
        return
    freq = dict(Counter(tokens).most_common(20))  # 20 palabras más frecuentes
    wc = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq)

    plt.figure(figsize=(8, 5))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(titulo)
    plt.show()

def visualizar_tsne(X_embeddings, y_labels, titulo="t-SNE de embeddings"):
    print("\nGenerando t-SNE...")

    tsne = TSNE(n_components=2, perplexity=30, max_iter=2000, random_state=42)
    X_2d = tsne.fit_transform(X_embeddings)

    df_vis = pd.DataFrame({
        "x": X_2d[:, 0],
        "y": X_2d[:, 1],
        "label": y_labels
    })

    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=df_vis, x="x", y="y", hue="label", palette="tab20", s=40)
    plt.title(titulo)
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

def visualizar_pca(X_embeddings, y_labels, titulo="PCA de embeddings"):
    print("\nGenerando PCA...")

    pca = PCA(n_components=2)
    X_2d = pca.fit_transform(X_embeddings)

    df_vis = pd.DataFrame({
        "x": X_2d[:, 0],
        "y": X_2d[:, 1],
        "label": y_labels
    })

    plt.figure(figsize=(12, 8))
    sns.scatterplot(data=df_vis, x="x", y="y", hue="label", palette="tab20", s=40)
    plt.title(titulo)
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

def validacion_cruzada(modelo, X, y, nombre="Modelo"):
    print(f"\nValidación cruzada (10-fold) para: {nombre}")

    cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

    scores_f1 = cross_val_score(modelo, X, y, cv=cv, scoring="f1_macro")
    scores_acc = cross_val_score(modelo, X, y, cv=cv, scoring="accuracy")
    scores_prec = cross_val_score(modelo, X, y, cv=cv, scoring="precision_macro")
    scores_rec = cross_val_score(modelo, X, y, cv=cv, scoring="recall_macro")

    resultados = {
        "Modelo": nombre,
        "Accuracy_mean": scores_acc.mean(),
        "Accuracy_std": scores_acc.std(),
        "F1_macro_mean": scores_f1.mean(),
        "F1_macro_std": scores_f1.std(),
        "Precision_macro_mean": scores_prec.mean(),
        "Recall_macro_mean": scores_rec.mean()
    }

    print("\nResultados (10-fold):")
    print(pd.DataFrame([resultados]))

    return resultados

def grid_search(modelo, params, X, y, nombre="Modelo"):
    print(f"\nGridSearchCV para: {nombre}")

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    grid = GridSearchCV(
        modelo,
        params,
        scoring="f1_macro",
        cv=cv,
        n_jobs=-1,
        verbose=1
    )

    grid.fit(X, y)

    print(f"\nMejores parámetros ({nombre}): {grid.best_params_}")
    print(f"Mejor puntaje F1-macro ({nombre}): {grid.best_score_:.4f}")

    return grid.best_estimator_, grid.best_params_, grid.best_score_

# Ejecutar
if __name__ == "__main__":
  print("\n===============================")
  print("              INICIO          ")
  print("===============================")

  # EDA
  df = eda(df)

  # Preprocesamiento
  df['cleaned'] = df['phrase'].astype(str).apply(preprocesamiento)
  print('\n', df[['phrase', 'cleaned']].head(), '\n')

  # MiniLM: generar embeddings
  X_mini = model_mini.encode(df['cleaned'].tolist(), convert_to_numpy=True)
  y_mini = df['prompt']

  # Visualizar solo las 10 clases más frecuentes
  top_clases = df['prompt'].value_counts().head(10).index
  filtro = df['prompt'].isin(top_clases)

  # Visualización t-SNE
  visualizar_tsne(X_mini[filtro], y_mini[filtro], "t-SNE de embeddings MiniLM (Sentence-BERT) - Top 10")
  # Visualización PCA
  visualizar_pca(X_mini[filtro], y_mini[filtro], "PCA de embeddings MiniLM (Sentence-BERT) - Top 10")

  # WordCloud general del corpus
  all_tokens = " ".join(df["cleaned"]).split()
  crear_wordcloud(all_tokens, "WordCloud general del corpus de síntomas médicos")

  # WordCloud por categoría (prompt)
  for clase in df['prompt'].unique():
    subset = df[df['prompt'] == clase]
    tokens = " ".join(subset['cleaned']).split()
    crear_wordcloud(tokens, f"WordCloud - {clase}")

  # Entrenamiento en TF-IDF
  X_entrena_tfidf, X_prueba_tfidf, y_entrena_tfidf, y_prueba_tfidf = entrenamiento(df)
  print('\nCOMO BASELINE\nDatos de entrenamiento TF-IDF:', X_entrena_tfidf.shape, y_entrena_tfidf.shape, 'Prueba TF-IDF:',  X_prueba_tfidf.shape, y_prueba_tfidf.shape)

  # Evaluación en TF-IDF
  resultados_tfidf = []

  # Evaluación en TF-IDF
  print("\nEvaluación con TF-IDF")
  for nombre, modelo in modelos.items():
      print(f"\n{nombre} (TF-IDF)")
      resultados_tfidf.append(
          evaluar_modelo(nombre, modelo,
                        X_entrena_tfidf, X_prueba_tfidf,
                        y_entrena_tfidf, y_prueba_tfidf)
      )

  # Pasar resultados a DataFrame
  resultados_tfidf_df = pd.DataFrame(resultados_tfidf)

  print("\n📊 Resultados TF-IDF:")
  print(resultados_tfidf_df)

  print("\n===============================")
  print("     EJECUCION GRID SEARCH     ")
  print("===============================")

  best_nb, params_nb, score_nb = grid_search(
      MultinomialNB(), param_grid_nb, X_entrena_tfidf, y_entrena_tfidf, "Naive Bayes"
  )

  best_svm, params_svm, score_svm = grid_search(
      SGDClassifier(random_state=42), param_grid_svm, X_entrena_tfidf, y_entrena_tfidf, "SVM"
  )

  best_lr, params_lr, score_lr = grid_search(
      LogisticRegression(max_iter=1000), param_grid_lr, X_entrena_tfidf, y_entrena_tfidf, "Logistic Regression"
  )

  print("\n===============================")
  print("  EJECUCION VALIDACION CRUZADA (10-fold)")
  print("===============================")

  resultados_cv = []

  for nombre, modelo in modelos.items():
      print(f"\nValidando {nombre}...")
      resultados_cv.append(
          validacion_cruzada(modelo, X_entrena_tfidf, y_entrena_tfidf, nombre)
      )

  # Convertir a DataFrame
  resultados_cv_df = pd.DataFrame(resultados_cv)

  print("\n📊 Resultados de Validación Cruzada (10-fold):")
  print(resultados_cv_df)

  # Tabla comparativa de modelos

  tabla_final = pd.DataFrame({
      "Modelo": ["Naive Bayes", "SVM", "Logistic Regression"],
      "TF-IDF F1_macro": [
          resultados_tfidf_df[resultados_tfidf_df["Modelo"] == "Naive Bayes"]["F1_macro"].values[0],
          resultados_tfidf_df[resultados_tfidf_df["Modelo"] == "SVM"]["F1_macro"].values[0],
          resultados_tfidf_df[resultados_tfidf_df["Modelo"] == "Logistic Regression"]["F1_macro"].values[0]
      ],
      "CrossVal F1_macro": [
          resultados_cv_df[resultados_cv_df["Modelo"] == "Naive Bayes"]["F1_macro_mean"].values[0],
          resultados_cv_df[resultados_cv_df["Modelo"] == "SVM"]["F1_macro_mean"].values[0],
          resultados_cv_df[resultados_cv_df["Modelo"] == "Logistic Regression"]["F1_macro_mean"].values[0]
      ],
      "GridSearch F1_macro": [score_nb, score_svm, score_lr],
      "Mejores parámetros": [params_nb, params_svm, params_lr]
  })

  print("\n=======================================")
  print("     TABLA COMPARATIVA FINAL DE MODELOS")
  print("=======================================\n")
  print(tabla_final.to_string(index=False))

  print("\n===============================")
  print("              FIN             ")
  print("===============================")
```
