```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout, GRU
from keras.optimizers import SGD
from keras import activations
import math
from sklearn.metrics import mean_squared_error

#%%
# get the data
path_file = '/content/drive/MyDrive/ClassFiles/Redes Neuronales/DataSets/GOOGL_2006-01-01_to_2018-01-01.csv'
dataset1 = pd.read_csv(path_file, index_col='Date', parse_dates=['Date'])
dataset1.head()

dataset = dataset1.dropna()  # Eliminar cualquier fila con NaN en cualquier columna

def preparar_datos(dataset, column="Low", previous_days=30):

  # Prepara los datos con la cantidad especificada

  training_set = dataset.loc[:'2016', [column]].values  # modificar dependiendo de la tarea
  test_set = dataset.loc['2017':, [column]].values   # modificar dependiendo de la tarea


  #%%
  # We have chosen 'Low' attribute for prices. Let's see what it looks like
  dataset[column][:'2016'].plot(figsize=(16,4),legend=True)
  dataset[column]['2017':].plot(figsize=(16,4),legend=True)
  plt.legend(['Training set (Before 2017)','Test set (2017 and beyond)'])
  plt.title('Google stock price')
  plt.show()

  #%%
  # Scaling the training set
  sc = MinMaxScaler(feature_range=(0,1))
  training_set_scaled = sc.fit_transform(training_set)

  # we create a data structure with 60 timesteps and 1 output
  # So for each element of training set, we have 60 previous training set elements
  X_train, y_train = [], []
  N = training_set.shape[0]
  for i in range(previous_days,N):
      X_train.append(training_set_scaled[i-previous_days:i,0])
      y_train.append(training_set_scaled[i,0])
  X_train, y_train = np.array(X_train), np.array(y_train)


  #%%
  # Reshaping X_train
  X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1],1))

  return X_train, y_train, test_set, sc

#%%

def crear_modelo(model='LSTM', input_shape=(30,1)):
  units = 50
  dropout = Dropout(0.2)

  # The architecture
  regressor = Sequential()
  RNNLayer = LSTM if model == 'LSTM' else GRU

  # First RNNLayer with Dropout regularisation
  regressor.add(RNNLayer(units=units, return_sequences=True, input_shape=input_shape))
  regressor.add(dropout)
  # Second LSTM layer
  regressor.add(RNNLayer(units=units, return_sequences=True))
  regressor.add(dropout)
  # Third LSTM layer
  regressor.add(RNNLayer(units=units, return_sequences=True))
  regressor.add(dropout)
  # Fourth LSTM layer
  regressor.add(RNNLayer(units=units))
  regressor.add(dropout)
  # The output layer
  regressor.add(Dense(units=1, activation='linear'))

  # Compiling the RNN
  regressor.compile(optimizer='rmsprop', loss='mean_squared_error')
  regressor.summary()

  return regressor


#%%

def predecir(regressor, dataset, sc, test_set, column='Low', previous_days=30):
  # get the test set ready in a similar way as the training set.
  dataset_total = pd.concat((dataset[column][:'2016'],dataset[column]['2017':]),axis=0)
  inputs = dataset_total[len(dataset_total)-len(test_set) - previous_days:].values
  inputs = inputs.reshape(-1,1)
  inputs  = sc.transform(inputs)

  # Preparing X_test and predicting the prices
  X_test = []
  N1 = inputs.shape[0]
  for i in range(previous_days,N1):
      X_test.append(inputs[i-previous_days:i,0])
  X_test = np.array(X_test)
  X_test = np.reshape(X_test, (X_test.shape[0],X_test.shape[1],1))

  #%%
  predicted_stock_price = regressor.predict(X_test)
  predicted_stock_price = sc.inverse_transform(predicted_stock_price)

  return predicted_stock_price

#%%

def visualizar(test_set, predicted_stock_price, company='Google', model='LSTM', days=30):
  # Visualizing the results for LSTM

  plt.plot(test_set, color='red',label='Real {} Stock Price'.format(company))
  plt.plot(predicted_stock_price, color='blue',label='Predicted {} Stock Price'.format(company))
  plt.title(f'{company} Stock Price Prediction with {model} ({days} days)')
  plt.xlabel('Time')
  plt.ylabel(f'{company} Stock Price')
  plt.legend()
  plt.show()

  # Evaluating our model
  rmse = math.sqrt(mean_squared_error(test_set, predicted_stock_price))
  print(f"RMSE para {model} a {days} días es {rmse}.")

  return rmse

#%%

resultados = []

for model in ['LSTM', 'GRU']:
  for days in [30, 50]:
    print(f"\nEntrenando {model} con {days} días...")
    X_train, y_train, test_set, sc = preparar_datos(dataset, column='Low', previous_days=days)
    reg = crear_modelo(model=model, input_shape=(days,1))
    reg.fit(X_train, y_train, epochs=10, batch_size=32)
    predicted = predecir(reg, dataset, sc, test_set, column='Low', previous_days=days)
    # Mostrar los primeros 10 valores predichos vs reales
    df_preview = pd.DataFrame({
        'Real': test_set.flatten()[:10],
        'Predicho': predicted.flatten()[:10]
    })
    print(df_preview)
    rmse = visualizar(test_set, predicted, company='GOOGL', model=model, days=days)
    resultados.append([model, days, rmse])

df_resultados = pd.DataFrame(resultados, columns=['Modelo','Días','RMSE'])
print("\nResultados finales:\n", df_resultados)
```
