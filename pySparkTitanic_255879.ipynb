##2. Creación de `SparkSession`

```python
# Creamos SparkSession, el punto de entrada para trabajar con DataFrames
from pyspark.sql import SparkSession

# Creación de la sesión de Spark
spark = (
    SparkSession.builder
    .appName("Titanic_DataFrame") # Nombre de la aplicación
    .master("local[*]") # Uso de todos los núcleos disponibles
    .getOrCreate()
)

print("SparkSession creada correctamente")
print("Versión de Spark:", spark.version)
```

##3. Carga de datos y Exploración inicial

```python
# Ruta del archivo en Google Drive previamente montado en Colab
df_titanic = spark.read.csv("/content/drive/MyDrive/ClassFilesIDA/Titanic-Dataset.csv", header=True, inferSchema=True)
# header=True indica que la primera fila contiene los nombres de las columnas
# inferSchema=True permite inferir automáticamente los tipos de datos

print("Dataset cargado correctamente")

# Visualización de las primeras 5 filas del DataFrame
df_titanic.show(5)

# Visualización del esquema del DataFrame (estructura y tipos de datos)
df_titanic.printSchema()
```

##4. Selección de columnas

```python
# Selección de las columnas Name, Age y Sex
df_select = df_titanic.select("Name", "Age", "Sex")

# Visualización del resultado de la selección
df_select.show(5)
```

##5. Filtrado de datos

```python
# Filtrado de pasajeros con las siguientes condiciones:
# - Edad mayor a 18 años
# - Sexo femenino
df_filtrado = df_titanic.filter((df_titanic["Age"] > 18) & (df_titanic["Sex"] == "female"))

# Visualización de los primeros 5 registros filtrados
df_filtrado.show(5)
```

##6. Valores únicos

```python
# Obtención de los valores únicos de la columna Pclass
df_titanic.select("Pclass").distinct().show()
```

##7. Agrupación
Por:
- Pclass
- Sex

```python
# Conteo del número de pasajeros por clase (Pclass)
df_titanic.groupBy("Pclass").count().show()

# Conteo del número de pasajeros por sexo
df_titanic.groupBy("Sex").count().show()
```

##8. Estadística básica

```python
# Cálculo de estadísticas descriptivas para las variables numéricas del DataFrame
df_titanic.describe().show()
```

##9. Valores nulos

```python
# Importación de funciones necesarias para identificar valores nulos
from pyspark.sql.functions import col, sum, when

# Identificación de la cantidad de valores nulos en cada columna
df_titanic.select([
    sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
    for c in df_titanic.columns
]).show()
```
